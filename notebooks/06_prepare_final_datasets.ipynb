```python
import pandas as pd

df = pd.read_csv("../data/processed/final_features_clean.csv")

df.shape
```

```python
ml_drop_cols = [
    "order_id",
    "customer_id",
    "product_id",
    "product_name",
    "delivery_partner_id",
    "store_id"
]
```

```python
df_ml = df.drop(columns=ml_drop_cols, errors="ignore")

df_ml.shape
```

```python
df_ml.to_csv(
    "../data/processed/final_features_ml.csv",
    index=False
)

print("✅ ML-ready dataset created")
```
